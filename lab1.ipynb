```python
import matplotlib.pyplot as plt
from collections import defaultdict

#read the file and return a dictionary with the sequence id as key and the sequence as value
def read_file(filename):
    with open(filename, "r") as file:
        lines = file.readlines()
    #initialize the dictionary
    seq = {}
    current_seq =""

    #iterate through the lines of the file
    for line in lines:
        line = line.strip()
        if line.startswith(">"): #if the line starts with >, = sequence id
            current_seq = line[1:]
            seq[current_seq] = ""
        else:
            seq[current_seq] += line.upper() #add the line to the sequence
    return seq

def count_letters(seq1): #count the number of each letter in the sequence
    counts = defaultdict(int)
    for letter in seq1:  #iterate through the sequence and count the number of each letter
        counts[letter] += 1
    return counts
#plot the frequency of each letter in the sequence
def dna_total(seq):
    for seq_id, seq1 in seq.items():
        counts = count_letters(seq1)
        letters = list(counts.keys())
        freq = list(counts.values())

        #set the font for the plot
        font1 = {'family': 'serif', 'color': 'black', 'weight': 'bold', 'size': 16}
        font2 = {'family': 'serif', 'color': 'darkred', 'weight': 'normal', 'size': 12}
        
        #plot the frequency of each letter
        plt.bar(letters, freq, color=['red', 'green', 'cyan', 'hotpink'])
        plt.ylabel("Freq", fontdict= font2)
        plt.xlabel("Letters", fontdict= font2)
        plt.title(f"Sequence: {seq_id}", fontdict = font1)
        plt.show()

#read the file and return a dictionary with the sequence id as key and the sequence as value
filename = "dna_raw_complicated.txt"
seq = read_file(filename)
dna_total(seq)

#read the file and return a dictionary with the sequence id as key and the sequence as value
filename = "dna_raw.txt"
seq2 = read_file(filename)
dna_total(seq2)
 
    
```
